```python
#Pandas is usually imported under the pd alias.
#Now the Pandas package can be referred to as pd instead of pandas
import pandas as pd
```

```python
# read_csv is an important pandas function to read csv files and do operations on it. 
df = pd.read_csv('tweets.csv', nrows=10000)
```

```python
#DataFrame.shape
# Return a tuple representing the dimensionality of the DataFrame.
df.shape
```

```python
#I use apply and lambda anytime I get stuck while building a complex logic for a new column or filter.
df['tweet_length'] = df['tweet'].apply(lambda x: len(x))
```

```python
sent = 'Hello, welcome to Pune. How is going...?'
```

```python
import string
```

```python
string.punctuation
```

```python
def count_punct(sent):
    count = sum([1 for x in sent if x in string.punctuation])
    p= round(count / (len(sent) - sent.count(' ')) * 100,2 )
    return p
```

```python
 count_punct(sent)
```

```python
df['punch%'] = df['tweet'].apply(lambda x: count_punct(x))
```

```python
df
```

```python
# a word stemmer based on porter stemming algorithm
# it is a process for removing commer morphological and inflexional ending for word in english
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
s_words = stopwords.words('english')
```

```python
def clean_text(text):
    data = [x for x in text if x not in string.punctuation]
    data = "".join(data)
    data = [ps.stem(x) for x in data.split()
           if x not in s_words]
    return data
    
```

```python
clean_text(sent)
```

```python
# input data
x = df.drop(['label','id'], axis=1)
# Ouput data
y = df['label']
```

```python
x
```

```python

from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(analyzer=clean_text)
x_tran = tfidf.fit_transform(x['tweet'])
```

```python
x_tran.shape
```

```python
x_vect = pd.concat([x[['tweet_length','punch%']]
                   .reset_index(drop=True),
                   pd.DataFrame(x_tran.toarray())], axis=1)
```

```python
x_vect
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_vect ,y, stratify=y, random_state =0 )
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression()
```

```python
clf.fit(X_train, y_train )
```

```python
y_pred = clf.predict(X_test)
```
